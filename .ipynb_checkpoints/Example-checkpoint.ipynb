```python
# Usual imports
import numpy as np
import xarray as xr
import pandas as pd
import polarplots as pp

# Load example data
ds=xr.open_dataset('pattern.nc', decode_times=False) 
mylat=ds['lat'].values
mylon=ds['lon'].values
myvar=ds['var151'][0,:,:].values

nrows=1
ncols=3

for i in range(3):
    if i==0:
        myfig=polaranom(mylat,mylon,myvar,clabel='colorbar label',
                        nrows=nrows, ncols=ncols, mapid=i+1)
    else:
        myfig=polaranom(mylat,mylon,myvar,figure=myfig,
                        nrows=nrows, ncols=ncols, mapid=i+1)

#myfig.savefig('panel.png', dpi=400)
myfig.savefig('panel_1x3.pdf')
```

```python
nrows=2
ncols=2

for i in range(4):
    if i==0:
        myfig=polaranom(mylat,mylon,myvar,clabel='colorbar label',
                        nrows=nrows, ncols=ncols, mapid=i+1)
    else:
        myfig=polaranom(mylat,mylon,myvar,figure=myfig,
                        nrows=nrows, ncols=ncols, mapid=i+1)

#myfig.savefig('panel.png', dpi=400)
myfig.savefig('panel_2x2.pdf')
```

```python
nrows=4
ncols=3

for i in range(12):
    if i==0:
        myfig=polaranom(mylat,mylon,myvar,lat0=30,clabel='colorbar label',
                        nrows=nrows, ncols=ncols,mapid=i+1)
    else:
        myfig=polaranom(mylat,mylon,myvar,lat0=30,figure=myfig,
                        nrows=nrows, ncols=ncols,mapid=i+1)

#myfig.savefig('panel.png', dpi=400)
myfig.savefig('panel_4x3.pdf')
```

```python
myfig=polaranom(mylat,mylon,myvar,lat0=30,
                colorbar=1,meridFontsize=8)

myfig.savefig('ex1.pdf')
```
